```python
import requests

url = 'https://raw.githubusercontent.com/SAURABH-RAI1729/KRITIGW/main/GW150914_strain_data_final.txt'

file = requests.get(url)

if file.status_code == 200:
    file = file.text
```

```python
lines = file.split('\n')
time_series_data = [line.split('\t') for line in lines if line]
```

```python
import numpy as np

data = np.array(time_series_data, dtype=object)
```

```python
time = data[:,0].astype(float)
value = data[:,1].astype(float)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(time, value, label='Scatter Plot')
plt.xlabel('Time')
plt.ylabel('Values')
plt.title('Scatter Plot of Time Series Data')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(time, value, label='Line Plot', linestyle = '-')
plt.xlabel('Time')
plt.ylabel('Values')
plt.title('Line Plot of Time Series Data')
plt.legend()
plt.grid(True)
plt.show()
```
